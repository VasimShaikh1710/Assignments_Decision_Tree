```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/admin/Desktop/Assignment_DataScience/Company_Data.csv')
df
```

```python
X = df.iloc[:,1:]
X
```

```python
Y = df['Sales']
Y
```

```python
mean = df['Sales'].mean()
mean
```

```python
from sklearn.preprocessing import Binarizer
bn = Binarizer(threshold=7.49)
Y = bn.fit_transform(df[["Sales"]])
Y
```

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers = [('tnf1', OrdinalEncoder(categories = [['Bad','Medium','Good']]),['ShelveLoc']),
                                                ('tnf2', OneHotEncoder(sparse = False, drop = 'first'),['Urban','US'])
                                                ], remainder = 'passthrough')
```

```python
trans_X = transformer.fit_transform(X)
trans_X
```

```python
New_X = pd.DataFrame(trans_X)
```

```python
New_X.columns = list(X)
New_X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(New_X,Y,train_size = 0.7,random_state = 24)
```

```python
from sklearn.tree import DecisionTreeRegressor
dc = DecisionTreeRegressor(max_depth = 4)
dc.fit(X_train,Y_train)
```

```python
Y_pred_train = dc.predict(X_train)
Y_pred_test = dc.predict(X_test) 
```

```python
print("Total number of node",dc.tree_.node_count)
```

```python
print("Total number of node",dc.tree_.max_depth)
```

```python
from sklearn.metrics import r2_score

print('Training Accuracy: ',r2_score(Y_train,Y_pred_train).round(2))
print('Test Accuracy: ',r2_score(Y_test,Y_pred_test).round(2))
```
